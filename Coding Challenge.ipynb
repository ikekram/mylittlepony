# Modeling

```python
X = df.loc[:, 'radius_mean' : 'fractal_dimension_worst']
y = df.diagnosis
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
X = std.fit_transform(X)
```

```python
from sklearn.cross_validation import train_test_split, KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, stratify=y)
```

```python
def evaluate_model(model):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)    
    a = accuracy_score(y_test, y_pred)    
    cm = confusion_matrix(y_test, y_pred)
    cr = classification_report(y_test, y_pred)    
    print cm
    print cr    
    return 'accuracy score', a
```

```python
#I'm getting hungry, time for a free lunch
import numpy
import xgboost as xgb
from sklearn import model_selection
from sklearn.grid_search import GridSearchCV


params = {
 'max_depth':range(3,8,1),
 'min_child_weight':range(1,8,2),
 'n_estimators':range(100,400,50)
}
gsearch1 = GridSearchCV(estimator = xgb.XGBClassifier( learning_rate =0.1, n_estimators=140, max_depth=5,
 gamma=0, subsample=0.8, colsample_bytree=0.8,
 objective= 'binary:logistic', nthread=4, scale_pos_weight=1, seed=27), 
 param_grid = params, scoring='accuracy',n_jobs=-1,iid=False, cv=KFold(len(y), n_folds=3, shuffle=True))
gsearch1.fit(X, y)
gsearch1.grid_scores_, gsearch1.best_params_, gsearch1.best_score_
```

```python
model = xgb.XGBClassifier( learning_rate =0.1, max_depth=4, gamma=0, subsample=0.8, colsample_bytree=0.8, objective= 'binary:logistic', nthread=4, scale_pos_weight=1, seed=27, min_child_weight=1, n_estimators=100)
evaluate_model(model)
```

### Model has 98.4% accuracy , that free lunch is delicious nomnomnom
